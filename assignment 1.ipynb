```python
import random
import string

def random_string(length):
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for _ in range(length))

for _ in range(100):
    length = random.randint(6, 8)
    print(random_string(length))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

prime_numbers = [num for num in range(600, 801) if is_prime(num)]
print("Prime numbers between 600 and 800:", prime_numbers)
```

```python
def div(n):
  if (n%7==0) and (n%9==0):
    return 1
div_by7and9=[num for num in range(100,1000) if div(num)]
print("div by 7 and 9 in range 100 to 1000:", div_by7and9)
```

```python
import random

# Generate two lists of random numbers
list1 = [random.randint(10, 30) for _ in range(10)]
list2 = [random.randint(10, 30) for _ in range(10)]

# Find common numbers in the two lists
common_numbers = set(list1).intersection(set(list2))

print("List 1:", list1)
print("List 2:", list2)
print("Common numbers:", list(common_numbers))
print("unique numbers:",set(list1).union(set(list2)))
print("maximum number:",max(max(list1),max(list2)))
print("minimum number:",min(min(list1),min(list2)))
```

```python
import random
import string

def generate_random_string(length):
    return ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(length))


with open("random_strings.txt", "w") as file:

    for _ in range(100):

        length = random.randint(600, 800)

        random_string = generate_random_string(length)

        file.write(random_string + "\n")
```
